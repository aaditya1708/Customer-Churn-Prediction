```python
df=df.dropna(subset="Churn")
```

```python
mask = df["DateOfJoining"].astype(str).str.match(r"^\d{8}$")

df.loc[mask, "DateOfJoining"] = pd.to_datetime(
    df.loc[mask, "DateOfJoining"],
    format="%Y%m%d"
)

df["DateOfJoining"] = pd.to_datetime(
    df["DateOfJoining"],
    dayfirst=True,
    errors="coerce"
)
```
